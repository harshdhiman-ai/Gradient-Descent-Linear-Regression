# Implement Gradient Descent for Linear Regression

```python


import numpy as np
from numpy.random import seed
from numpy.random import randint
seed(1)

global eta  ## Learning rate 
eta=0.001


def gradient_descent(x,y): 
    w0=w1=0   # Initialize the weights w1 and w0 as per equation y=w1*x+w0
    itr=100 # number of iterations for which weights are updated
    n=len(x)    
    for i in range(itr):
        ycap=w1*x+w0
        cost=1/n*sum([val**2 for val in (y-ycap)])
        J1=-2/n*sum(x*(y-ycap))
        J2=-2/n*sum(y-ycap)
        
        w1=w1-eta*J1
        w0=w0-eta*J2
        
        print("cost {}, w1 {}, w0 {}, itr {}".format(cost,w1,w0,i)) 
        
x=np.array([1,4.5,5,7.88,6.134,51])
y=np.array([3,13,14.77,23,31,67])



gradient_descent(x,y)
```
